```python
import pandas as pd

global_happiness = pd.read_csv("global_happiness_df.csv",  index_col=0)
hdi = pd.read_csv("HDIdataset.csv",  index_col=0)
quality_of_life = pd.read_csv("qol_df.csv",  index_col=0)

print("Global Happiness columns:", global_happiness.columns.tolist())
print("HDI columns:", hdi.columns.tolist())
print("Quality of Life columns:", quality_of_life.columns.tolist())

countries_happiness = set(global_happiness["country"].unique())
countries_hdi = set(hdi["country"].unique())
countries_qol = set(quality_of_life["Country"].unique())

print(f"\nGlobal Happiness Countries ({len(countries_happiness)}):")
print(sorted(countries_happiness))

print(f"\nHDI Countries ({len(countries_hdi)}):")
print(sorted(countries_hdi))

print(f"\nQuality of Life Countries ({len(countries_qol)}):")
print(sorted(countries_qol))
```

```python
diff_hap_hdi = countries_happiness - countries_hdi
diff_hdi_hap = countries_hdi - countries_happiness
print(f"\nCountries in Global Happiness not in HDI ({len(diff_hap_hdi)}):")
print(sorted(diff_hap_hdi))
print(f"\nCountries in HDI not in Global Happiness ({len(diff_hdi_hap)}):")
print(sorted(diff_hdi_hap))
```

```python
diff_hap_qol = countries_happiness - countries_qol
diff_qol_hap = countries_qol - countries_happiness
print(f"\nCountries in Global Happiness not in Quality of Life ({len(diff_hap_qol)}):")
print(sorted(diff_hap_qol))
print(f"\nCountries in Quality of Life not in Global Happiness ({len(diff_qol_hap)}):")
print(sorted(diff_qol_hap))
```

```python
def align_country_names(name):
    mapping = {
        # Alignments from Global Happiness to HDI / Quality of Life names
        "Bolivia": "Bolivia (Plurinational State of)",
        "Congo (Brazzaville)": "Congo (Democratic Republic of the)",
        "Congo (Kinshasa)": "Congo (Democratic Republic of the)",
        "Czech Republic": "Czechia",
        "Eswatini": "Eswatini (Kingdom of)",
        "Hong Kong S.A.R. of China": "Hong Kong",
        "Iran": "Iran (Islamic Republic of)",
        "Ivory Coast": "Côte d'Ivoire",
        "Laos": "Lao People's Democratic Republic",
        "Moldova": "Moldova (Republic of)",
        "North Cyprus": "Northern Cyprus",
        "Palestinian Territories": "Palestine, State of",
        "Russia": "Russian Federation",
        "Somaliland Region": "Somaliland Region",
        "Somaliland region": "Somaliland Region",
        "South Korea": "Korea (Republic of)",
        "State of Palestine": "Palestine, State of",
        "Swaziland": "Eswatini (Kingdom of)",
        "Syria": "Syrian Arab Republic",
        "Taiwan Province of China": "Taiwan",
        "Tanzania": "Tanzania (United Republic of)",
        "Turkey": "Türkiye",
        "Turkiye": "Türkiye",
        "Venezuela": "Venezuela (Bolivarian Republic of)",
        "Vietnam": "Viet Nam",
        # Alignments for Quality of Life differences
        "Bosnia And Herzegovina": "Bosnia and Herzegovina",
        "Hong Kong (China)": "Hong Kong",
        "Taiwan (China)": "Taiwan",
    }
    return mapping.get(name, name)


global_happiness["country"] = global_happiness["country"].apply(align_country_names)
hdi["country"] = hdi["country"].apply(align_country_names)
quality_of_life["Country"] = quality_of_life["Country"].apply(align_country_names)
```

```python
quality_of_life = quality_of_life.rename(columns={"Country": "country"})

# Extract unique, aligned country names.
countries_happiness_aligned = set(global_happiness["country"].unique())
countries_hdi_aligned = set(hdi["country"].unique())
countries_qol_aligned = set(quality_of_life["country"].unique())

diff_hap_hdi = countries_happiness_aligned - countries_hdi_aligned
diff_hdi_hap = countries_hdi_aligned - countries_happiness_aligned
print("\nAfter alignment, Countries in Global Happiness not in HDI ({}):".format(len(diff_hap_hdi)))
print(sorted(diff_hap_hdi))
print("\nAfter alignment, Countries in HDI not in Global Happiness ({}):".format(len(diff_hdi_hap)))
print(sorted(diff_hdi_hap))

diff_hap_qol = countries_happiness_aligned - countries_qol_aligned
diff_qol_hap = countries_qol_aligned - countries_happiness_aligned
print("\nAfter alignment, Countries in Global Happiness not in Quality of Life ({}):".format(len(diff_hap_qol)))
print(sorted(diff_hap_qol))
print("\nAfter alignment, Countries in Quality of Life not in Global Happiness ({}):".format(len(diff_qol_hap)))
print(sorted(diff_qol_hap))
```

```python
global_happiness["Year"] = global_happiness["Year"].astype(str)
hdi["Year"] = hdi["Year"].astype(str)
quality_of_life["Year"] = quality_of_life["Year"].astype(str)

merged_df = pd.merge(global_happiness, hdi, on=["country", "Year"], how="outer", suffixes=("_happy", "_hdi"))
merged_df = pd.merge(merged_df, quality_of_life, on=["country", "Year"], how="outer")
merged_df.head()
```
